Activation Functions are used in machine learning to describe complicated interactions between inputs and outputs, and activation functions are a crucial part of these networks. The network can learn more complex functions by adding non-linearity through activation functions.

```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(-10, 10, 2)
print(x)
```

```python
def plot_graph(y,ylabel):
    plt.figure()
    plt.plot(x,y, 'o--')
    plt.grid(True)
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()
```

```python
y = list(map(lambda n: 1 if n>0.5 else 0, x))
plot_graph(y,"STEP(X)")
```

1.STEP FUNCTION

The step activation function is a simple type of activation function utilized in artificial neural networks, generally for binary classification issues, is the step activation function. Depending on whether the input value exceeds or falls short of a predetermined threshold value, it converts the input value to a binary output value, either 0 or 1.

The output of the step function is discontinuous, abruptly changing from 0 to 1 as soon as an input value reaches a threshold. Due to the fact that the derivative of the step function is almost always 0, it is less ideal for training neural networks using gradient-based optimization techniques. This makes it challenging to update the network's weights during backpropagation.

```python
y = 1 / (1 + np.exp(-x))
plot_graph(y, "Sigmoid(X)")
```

2.SIGMOID FUNCTION

It is use to add nonlinearity to a node's output is the sigmoid activation function. Any input value is translated into a value between 0 and 1 by a smooth, S-shaped function.

There are a few significant characteristics of the sigmoid function that are advantageous for neural networks. It has a derivative at every point since, to start, it is a continuous function. This enables the neural network to be trained using gradient-based optimization methods. Second, because it is a non-linear function, it can add non-linearities into the neural network's output, enabling it to discover intricate patterns in the data.

```python
y = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
plot_graph(y, "TanH(X)")
```

3.TANH FUNCTION

It provides nonlinearity to a node's output is the hyperbolic tangent (tanh) activation function. It is comparable to the sigmoid function and is beneficial for problems where negative values are important since it transfers input values to a range between -1 and 1.

The tanh function is continuous and non-linear, just as the sigmoid function, making it helpful for neural networks. Also, it has the characteristic of being zero-centered, or having an output that is centered on zero, which can aid in the network's convergence during training.

```python
y = list(map(lambda a: a if a>=0 else 0, x))
plot_graph(y,"ReLU(X)")
```

4.RELU - RECTIFIED LINEAR UNITS

The rectified linear unit (ReLU) activation function is a popular mathematical function used in artificial neural networks to provide nonlinearity to a node's output is the rectified linear unit (ReLU) activation function. It is a straightforward function that, if the input value is positive, returns the value; else, it returns zero.

The ReLU function has various advantages over other activation functions. It is a straightforward function that can be computed quickly. Second, it doesn't have the issue with vanishing gradients for big input values, which makes deep neural network training simpler. Thirdly, it can aid in sparsifying the network's activation by setting some activations to zero, which can aid in avoiding overfitting.

```python
alpha = 0.1

y = list(map(lambda a: a if a>=0 else a*0.1, x))
plot_graph(y,"ReLU(X)")
```

5.PRELU

When compared to other activation functions, the PReLU function has a number of benefits. First, by allowing for non-zero output for negative input values, it can aid in preventing the issue of dying ReLU. Second, it can aid in increasing the network's accuracy, particularly for datasets with a lot of noise or fluctuation. Thirdly, it is a straightforward function that computes quickly.

```python
import numpy as np

def eReLU(x, alpha=0.01):
    """ Exponential Rectified Linear Unit activation function """
    return np.where(x > 0, x, alpha*(np.exp(x) - 1))

x = np.linspace(-5, 5, 40)
y = list(map(lambda a: eReLU(a), x))

plot_graph(y, "eReLU(X)")
```

6.ERELU

A variant of the well-known ReLU activation function used in artificial neural networks is the Exponential Rectified Linear Unit (eReLU) activation function. The "dying ReLU" problem, which happens when ReLU neurons with negative inputs get "stuck" at 0 and cease learning, is addressed by the eReLU function.
